```python
import numpy as np
import pandas as pd

def perceptron(x, w, b):

    # cálculo del puntaje
    z = np.dot(x, w) + b

    # función de activación
    if z >= 0:
        salida = 1
    else:
        salida = 0

    return z, salida
    # pesos
w = np.array([1,1])

# bias
b = -1
pruebas = [
    [0,0],
    [0,1],
    [1,0],
    [1,1]
]

resultados = []

for p in pruebas:

    x = np.array(p)

    z, salida = perceptron(x, w, b)

    resultados.append([p[0], p[1], z, salida])

    print("Entrada:", p, "z:", z, "Salida:", salida)
    tabla = pd.DataFrame(resultados, columns=["x1","x2","z","Salida"])

tabla
```

El perceptrón calcula un puntaje z utilizando las entradas, los pesos y el bias.

z = (x1*w1) + (x2*w2) + bias

Si z es mayor o igual a 0 la salida es 1, de lo contrario es 0.

En las pruebas realizadas se observa que cuando al menos una de las entradas es 1, el valor de z aumenta y la salida del modelo cambia.

El bias permite desplazar el punto de decisión del modelo. Si se modifica el bias, el comportamiento del perceptrón cambia y puede clasificar los datos de forma diferente.
